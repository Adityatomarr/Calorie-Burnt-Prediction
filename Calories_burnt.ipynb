Converting the test data to numerical data

```python
calories_data.replace({"Gender": {'male':0, 'female':1}}, inplace=True)
```

```python
calories_data.head()
```

Separating features and target

```python
X= calories_data.drop(columns=['User_ID', 'Calories'], axis=1)
Y= calories_data['Calories']
```

```python
print(X)
```

```python
print(Y)
```

Code to split training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state =2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor

```python
#loading the model
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on test data

```python
test_data_prediction = model.predict(X_test)
```

```python
print(test_data_prediction)
```

Mean Absolute error

```python
 mae = metrics.mean_absolute_error(Y_test, test_data_prediction)
```

```python
print("Mean absolute error = ", mae)
```

Catboost

---



```python
modelCat = cb.CatBoostRegressor(loss_function='MAE' , iterations =500)
modelCat.fit(X_train , Y_train)
```

Prediction

```python
#predicting data for catboost
pred = modelCat.predict(X_test)
```

```python
print(pred)
```

```python
mae2 = metrics.mean_absolute_error(Y_test, pred)
print("Mean absolute error = ", mae2)
```
